```python
epldata_final_df = pd.read_csv('.\datasets\epldata_final.csv')
epldata_final_df
```

```python
df_3 = pd.read_csv('.\datasets\WA_Fn-UseC_-Telco-Customer-Churn.csv', index_col = 0)
df_3.head()
```

```python
df_3[df_3.duplicated()]
```

```python
df_no_duplicate = df_3.drop_duplicates()
df_no_duplicate.info()
```

```python
pd.get_dummies(df_no_duplicate['TotalCharges'])
```

```python
df_no_duplicate['TotalCharges'] = pd.to_numeric(df_no_duplicate['TotalCharges'], errors='coerce')
```

```python
df_no_duplicate.info()
```

```python
df_no_duplicate['TotalCharges'].mean()
```

```python
fill_values = {'TotalCharges':df_no_duplicate['TotalCharges'].mean()}
df_no_duplicate = df_no_duplicate.fillna(value = fill_values)
df_no_duplicate.info()
```

```python
pd.get_dummies(df_no_duplicate[['gender', 'InternetService','PaymentMethod']])
```

```python
df_encoding = pd.concat([pd.get_dummies(df_no_duplicate[['gender', 'InternetService','PaymentMethod']]),df_no_duplicate],axis=1)
df_encoding
```

```python
df_encoding.drop(['gender','InternetService','PaymentMethod'], axis=1, inplace=True)
df_encoding
```

```python
encoding_map = {"Month-to-month":1 ,"One year":2 , "Two year":3}
df_encoding['Contract'] = df_encoding['Contract'].replace(encoding_map)
df_encoding
```

```python
df_encoding['Contract'] = pd.to_numeric(df_encoding['Contract'],errors= 'coerce')
df_encoding.info()
```

```python
ordinal_label_map = {'Yes':1 , 'No':0}
df_encoding['Churn'] =  df_encoding['Churn'].replace(ordinal_label_map)
```

```python
df_encoding
```

```python
Target_labels = df_encoding.loc[:,'Churn']
type(Target_labels)
```

```python
Target_labels = pd.DataFrame(Target_labels)
Target_labels
#type(Target_labels)
```

```python
#del df_encoding['Churn']
df_encoding
```

```python
 pd.get_dummies(df_encoding[['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity', 'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling']])
```

```python
df_final = pd.concat([df_encoding,pd.get_dummies(df_encoding[['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity', 'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling']])], axis=1)
df_final.info()
```

```python
df_final.drop(['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity', 'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling'] ,axis=1, inplace=True)
```

```python
df_final.info()
```

```python
normalized_final_df = (df_final - df_final.min())/(df_final.max()-df_final.min())
```

```python
normalized_final_df
```

```python
normalized_final_df.to_csv('written_normalized_final_df.csv', index=True, sep=",")
```
